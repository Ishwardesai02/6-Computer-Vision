```python
import cv2
#Connects to your Computer's camera
cap=cv2.VideoCapture(0)



#Automatically grab width and height from video feed
#(returns float which we need to covert to integer for later on!)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#We're using // foo flour division

#coordinates for rectangle , center point rectangle
x=width//2
y=height//2

#width and height adding 1/4 of width and height from the center
w=width//4
h=height//4


#This loop keep recording until you hit 0 or escape the window
while True:
    #capture frame by frame
    ret,frame=cap.read()
    
    #draw rectangle on stream
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=4)
 
    
    #Display the resulting frame
    cv2.imshow('frame',frame)
    
    #This command let's us quit with "q" button on a keyboard.
    #Simply pressing X on the window won't work
    if cv2.waitKey(1) & 0xff==ord('q'):
        break
        
#When everything done,release the capture and destroy the windows
cap.release()
cv2.destroyAllWindows()

    
```

```python
import cv2
import numpy as np

# Global variables
pt1 = (0, 0)
pt2 = (0, 0)
topLeft_clicked = False
botRight_clicked = False

# Mouse callback function to reset the rectangle
def draw_rectangle(event, x, y, flags, param):
    global pt1, pt2, topLeft_clicked, botRight_clicked

    # Check for left mouse button down event
    if event == cv2.EVENT_LBUTTONDOWN:
        # Reset the rectangle if both corners have been clicked
        if topLeft_clicked and botRight_clicked:
            topLeft_clicked = False
            botRight_clicked = False
            pt1 = (0, 0)
            pt2 = (0, 0)

        # Set the first point if the top-left corner hasn't been clicked yet
        if not topLeft_clicked:
            pt1 = (x, y)
            topLeft_clicked = True
        # Set the second point if the top-left corner is already clicked
        elif not botRight_clicked:
            pt2 = (x, y)
            botRight_clicked = True

# Capture video from webcam
cap = cv2.VideoCapture(0)

# Create a named window for the connection
cv2.namedWindow('Test')

# Bind the draw_rectangle function to mouse clicks
cv2.setMouseCallback('Test', draw_rectangle)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Draw a circle at the first point
    if topLeft_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0, 0, 255), thickness=-1)

    # Draw the rectangle if both points are clicked
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)

    # Display the resulting frame
    cv2.imshow('Test', frame)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
